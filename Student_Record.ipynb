```python
import pandas as pd
import os
```

```python
import kagglehub
path = kagglehub.dataset_download("pratikprasad18/student-records")
```

```python
record=pd.read_csv(os.path.join(path,"student_records.csv"))
```

```python
record
```

```python
record=record.drop(['AttendancePercent'],axis=1)
```

```python
record
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
coder=LabelEncoder()
```

```python
record['StudentID']=record['StudentID'].astype(str)
record['StudentID']=coder.fit_transform(record['StudentID'])
```

```python
record['Name']=record['Name'].astype(str)
record['Name']=coder.fit_transform(record['Name'])
```

```python
record['Gender']=record['Gender'].astype(str)
record['Gender']=coder.fit_transform(record['Gender'])
```

```python
record['Extracurricular']=record['Extracurricular'].astype(str)
record['Extracurricular']=coder.fit_transform(record['Extracurricular'])
```

```python
record['Grade']=record['Grade'].astype(str)
record['Grade']=coder.fit_transform(record['Grade'])
```

```python
record
```

```python
x=record.drop(['Grade'],axis=1)
y=record['Grade']
```

```python
y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import seaborn as seb
plt.scatter(record['Math'], record['Science'])
plt.xlabel('Math')
plt.ylabel('Science')
plt.title('Scatter Plot of Math vs Science')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_test.info()
```

```python
x_train.info()
```

```python
y_test.info()
```

```python
y_train.info()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn import metrics
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))
print('R-squared:', metrics.r2_score(y_test, predictions))
```

```python
import pandas as pd

# Create a DataFrame to compare actual vs. predicted grades
predictions_df = pd.DataFrame({
    'Actual Grade': y_test,
    'Predicted Grade': predictions
})

# Display the first few comparisons
print("Sample of Actual vs. Predicted Grades:")
display(predictions_df.head(10))
```

```python
from sklearn.metrics import classification_report, accuracy_score
import numpy as np


min_grade = y_test.min()
max_grade = y_test.max()


predicted_grades_discrete = np.round(predictions)
predicted_grades_discrete = np.clip(predicted_grades_discrete, min_grade, max_grade).astype(int)

print("Classification Report:")
print(classification_report(y_test, predicted_grades_discrete))
print("\nAccuracy Score:", accuracy_score(y_test, predicted_grades_discrete))
```
